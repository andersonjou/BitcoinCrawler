# Bitcoin Crawler and Notification 

Using BeautifulSoup and Pandas to crawl bitcoin's price from coingecko.com and send notification via E-mail.

## Part 1 - Import Data

```python
import pandas as pd
import io
import requests
url = "https://www.coingecko.com/price_charts/export/1/usd.csv"
s = requests.get(url).content
df = pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
import matplotlib as plt
%matplotlib inline
```

```python
df
```

```python
df['time'] = pd.to_datetime(df['snapped_at'])
```

```python
df.index = df["time"]
```

```python
df
```

```python
df = df.drop(columns="snapped_at")
df = df.drop(columns="time")
```

## Part 3 - Technical analysis

```python
# calculate MA, EMA
ma_list = [12, 26]
for ma in ma_list:
    df['EMA_' + str(ma)] = df["twd"].ewm(span = ma).mean()
for ma in ma_list:
    df['MA_' + str(ma)] = df["twd"].rolling(window = ma).mean()
    
```

```python
# calculate MA, EMA
ma_list = [12, 26]
for ma in ma_list:
    df2['EMA_' + str(ma)] = df2["price"].ewm(span = ma).mean()
for ma in ma_list:
    df2['MA_' + str(ma)] = df2["price"].rolling(window = ma).mean()
```

## Why? 

```python
df3 = df2.copy()
```

```python
df3[["price", "EMA_26", "MA_26"]].plot(figsize = (15, 6))
```

```python
# calculate DIF, DEM
df3['dif'] = df3['EMA_12'] - df3['EMA_26']
df3["dem"] = df3["dif"].ewm(span = 9).mean()

# plot  for Moving Average Convergence / Divergence
df3[['price']].plot(kind = "line", figsize = [20, 5], fontsize = 10)
df3[['dif','dem']].plot(kind = "line", figsize = [20,5], fontsize = 10)
```
