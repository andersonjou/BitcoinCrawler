## Part 3 - Technical analysis

```python
# calculate MA, EMA
ma_list = [12, 26]
for ma in ma_list:
    df2['EMA_' + str(ma)] = df2["price"].ewm(span = ma).mean()
for ma in ma_list:
    df2['MA_' + str(ma)] = df2["price"].rolling(window = ma).mean()
```

```python
df2
```

## Why? 

### 3-1 Moving Average

```python
df3 = df2.copy()
```

```python
df3[["price", "EMA_26", "MA_26"]].plot(figsize = (15, 6))
```

### 3-2 MACD

```python
# calculate DIF, DEM
df3['dif'] = df3['EMA_12'] - df3['EMA_26']
df3["dem"] = df3["dif"].ewm(span = 9).mean()

# plot  for Moving Average Convergence / Divergence
df3[['price']].plot(kind = "line", figsize = [20, 5], fontsize = 10)
df3[['dif','dem']].plot(kind = "line", figsize = [20,5], fontsize = 10)
```

### 3-3 Bollinger bands

```python
import numpy as np
```

```python
df3["MA_20"] = df3["price"].rolling(window = 20).mean()
```

```python
df3["upper_band"] = df3["MA_20"] + 2 * df3["price"].rolling(window = 20).std() 
```

```python
df3["lower_band"] = df3["MA_20"] - 2 * df3["price"].rolling(window = 20).std() 
```

```python
df3
```

```python
df3[["price", "upper_band", "lower_band"]].plot(figsize = (15, 6))
```

```python
def boll_detecter(today_price, today_upper, today_lower):
    if (today_price >= today_upper) == True:
        return("Sell")
    elif (today_price >= today_lower and today_price < today_upper) == True:
        return("Hodl!")
    elif (today_price < today_lower) == True:
        return("Buy")
```

```python
boll_detecter(today_price, today_upper, today_lower)
```
